```python
# specify the location of the file you wish to take a look at
datafile_path = "../_output/8bb738d2-6954-11ea-a95f-185680feb82e.data.json"
```

```python
import json

# open the datafile
with open(datafile_path, 'r') as datafile:
    data = json.load(datafile)
```

```python
import graphing_utilities as gutil

# graph it using the example graphing utility
gutil.plot_data(data)
```

```python
import os

# we can save the figure if we give it a folder
figure_folder = "../_output/figures/"

os.makedirs(figure_folder, exist_ok=True)  # but it needs to exist
"""exist_ok=True means that it won't overwrite it if it already exists"""

gutil.plot_data(data, save_dir=figure_folder)
```
